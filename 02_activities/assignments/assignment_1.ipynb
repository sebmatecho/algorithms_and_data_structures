```python
import hashlib

def hash_to_range(input_string: str) -> int:
     hash_object = hashlib.sha256(input_string.encode())
     hash_int = int(hash_object.hexdigest(), 16)
     return (hash_int % 3) + 1
input_string = "sebastien"
result = hash_to_range(input_string)
print(result)
```

  You are given a list containing `n` integers in the range `[0, n]`. Return a list of numbers that are missing from the range `[0, n]` of the array. If there is no missing number, return -1. Note, all the integers in the list may not be unique.
  
  ## Examples

  Output: [1]

  ### Example 2

  Input: `lst = [5, 0, 1]`

#### Starter Code for Question 3


```python
from typing import List, Union

def missing_num(nums: List[int]) -> Union[int, List[int]]:
    n = max(nums)
    
    # baseline of numbers
    baseline_nums = set(range(n + 1))

    # input into set
    nums_set = set(nums)
    
    # checking the difference 
    missing = list(baseline_nums - nums_set)
    
    # If there are no missing numbers, return -1
    if not missing:
        return -1
    else: 
        return missing
```

```python
print(missing_num(nums = [0, 2]))
print(missing_num(nums = [5, 0, 1]))
print(missing_num(nums = [6, 8, 2, 3, 5, 7, 0, 1, 10]))
```

```python
from typing import List, Union

def missing_num(nums: List[int]) -> Union[int, List[int]]:
    n = max(nums)
    
    # baseline of numbers
    baseline_nums = set(range(n + 1))

    # input into set
    nums_set = set(nums)
    
    # checking the difference 
    missing = list(baseline_nums - nums_set)
    
    # If there are no missing numbers, return -1
    if not missing:
        return -1
    else: 
        return missing
```


-   Explain why your solution works


my solution goes: 
- for the largest value contained in the input, say $n$ 
- goes ahead to create the baseline set containing all integers in $0, \cdots, n$  
- compares such baseline set with the given input to spot differences
- such differences are stored and returned
- if no differences are found, returns -1


-   Explain the problem’s time and space complexity


Time Complexity ($O(n)$)
- Finding max(nums) takes $O(n)$ as it ranges through all numbers
- Creating sets for comparison takes $O(n)$ as it creates all numbers
- Computing the set difference takes $O(n)$ as it ranges through all numbers
- Converting the missing elements back into a list takes $O(m)$ (where $m$ is the number of missing elements, at most $n$).
- Since $O(m) \leq O(n)$, the overall time complexity is O(n).

Space Complexity $(O(n))$
- A set of all numbers (baseline_nums) takes $O(n)$.
- A set of existing numbers (nums_set) takes $O(n)$.
- The list of missing numbers (missing) takes $O(m)$.
- Since $m \leq n$, the worst-case space complexity is $O(n)$.
