
-   Copy the solution your partner wrote. 


```python
# Your answer here
# Definition for a binary tree node.
class TreeNode(object):
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

def is_duplicate(root: TreeNode) -> int:
    seen = set()
    queue = [root]
    while queue:
        node = queue.pop(0)
        if node.val in seen:
            return node.val
        seen.add(node.val)
        if node.left:
            queue.append(node.left)
        if node.right:
            queue.append(node.right)
    return -1
```


-   Explain why their solution works in your own words.


```python
# Your answer here
# I imagine my partner inspecting the tree level by level, starting at the root. He keeps a set of values he has already encountered. 
# For each node he looks at, he checks whether its value is already in the set. If it is, that means he has found a duplicate, so he stops and returns such value. 
# If it isn’t, he adds the value to the set and then adding the node’s children to the “to-do” list. 
# This “to-do” list is actually a queue, which guarantees that he processes nodes in the order of their distance from the root. 
# Once the end of the tree is reached without finding any duplicate, -1 is to be returned.
```


-   Explain the problem’s time and space complexity in your own words.


```python
# Your answer here
# Time Complexity: 𝑂(𝑛) where n is the number of nodes. Each node is enqueued and dequeued exactly once, and set membership checks (in seen) are O(1) on average.
# Space Complexity: O(n) in the worst case, for both the queue and the seen set, if almost every node is added before finding a duplicate or if no duplicate exists.
```


-   Critique your partner's solution, including explanation, and if there is anything that should be adjusted.
